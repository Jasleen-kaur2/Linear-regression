```python
##train test and learn- websites-
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x,y,random_state=42,test_size=0.25) 
#random_state-cycle tht goes till the 80:20 ratio is acieved,
#test_size-it divides the model in 80:20 test and train
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_train
```

```python
x_test=scaler.transform(x_test)
x_test
```

```python
#made regression model
from sklearn.linear_model import LinearRegression
regression=LinearRegression(n_jobs=-1)
```

```python
regression.fit(x_train,y_train)
regression
```

```python
##prediction- value came from out model is stored in y_pred
y_pred=regression.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)#comparin gvalue that came with value that should come
print(score)
```
